# Fluxonium qubit

$H_\text{fl}=-4E_\text{C}\partial_\phi^2-E_\text{J}\cos(\phi-\varphi_\text{ext}) +\frac{1}{2}E_L\phi^2$

**Creation via GUI** (ipywidgets needs to be installed for this to work.)

```python
fluxonium = scq.Fluxonium.create()
```

```python
print(fluxonium)
```

**Programmatic creation**

```python
fluxonium2 = scq.Fluxonium(
    EJ=6.25,
    EC=0.75,
    EL=0.91,
    cutoff = 110,
    flux = 0.5
)
```

## Computing and visualizing spectra

```python
fluxonium.eigenvals()
```

```python
fluxonium.eigenvals()
```

```python
flux_list = np.linspace(0.41, 0.6, 151)
fluxonium.plot_evals_vs_paramvals('flux', flux_list,subtract_ground=1 ,evals_count=4);
#plt.ylim([0.06,0.08])
```

```python
fluxonium.cos_phi_operator()
```

```python
flux_list = np.linspace(0.44, 0.5, 151)
fluxonium.plot_evals_vs_paramvals('flux', flux_list, evals_count=8);
```

```python
fluxonium.flux = 0.5
fluxonium.plot_wavefunction(esys=None, which=range(5), mode='imag');
```

```python
fluxonium = scq.Fluxonium(
    EJ=9,
    EC=0.75,
    EL=0.91,
    cutoff = 30,
    flux = 0.45
)
fig, axes=fluxonium.plot_wavefunction(esys=None, which=[0,1,2,3,4,5], mode='real');
```

## Matrix elements

```python
phimat = fluxonium.matrixelement_table('phi_operator', evals_count=10)
print(phimat)
```

```python
fluxonium.plot_matrixelements('n_operator', evals_count=16);
```

```python
fig, axes = fluxonium.plot_matelem_vs_paramvals('n_operator', 'flux', flux_list, select_elems=[(1, 3),(0,2)]);
labelLines(axes.get_lines(), xvals=(0.1, 1), zorder=2.5)
```
